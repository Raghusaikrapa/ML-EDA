# NP.WHERE

- np.where required 3 parameters
- condation
- True
- False
- np.where is applicable for only binary(0,1)
- example in case status where have 2 lables (certified,dined)
- so if visa_df['case_status']=='Certified' then =1 else=0

**NOTE:** this is aplicable if only two lables are avalible in a column

```python
visa_df=pd.read_csv('Visadataset - Visadataset.csv')
visa_df
```

```python
con=visa_df['case_status']=='Certified'
then=0
orthen=1
visa_df['case_status']=np.where(con,then,orthen)
```

# ONE HEART ENCODER

- pd.getdummyes

- as name suggests one heart encoder if one on other will off
- for example case_status has 2 columns
  - Certified
  - Denied
- here one heart encoder will create 2 other columns
  - as case_status_certified
  - case_status_Denied

```python
Case_status     Case_status_certified         Case_status_denied

certified              1                              0

denied                 0                              1
```

**Advantages**

- when we are deveplping the ML model it is very importent that varables are independent to eachother
- here case_status creacted 2 seperated columns
- thus 2 are independent to each other at a time onely one column has value === 1
- columns are independent to eachother
- which means 90 degrees phas shift
- which means perpendicular to eachother
- which means orthogonal to eachother

**Distavantages**

- the main disadvantage is which we have 100 unique lables in a column it created 100 columns
- takes more time for exicuation
- takes more memory
- data becomes huge
- if columns are more then dimentations are more
