- DO THE SAME FOR REPLACING MEDIAN VALUES

```python
lb=-76564.56875000002
ub=218315.56125000003
```

```python
md=visa_df['prevailing_wage'].median()
o1=visa_df['prevailing_wage']<lb
o2=visa_df['prevailing_wage']>ub
con=o1 | o2
t=md
f=visa_df['prevailing_wage']
```

```python
visa_df['prevailing_wage_new']=np.where(con,t,f)
visa_df
```

```python
# NOW LETS PLOT BOXPLOT TO CHECK outlayes present or not
import matplotlib.pyplot as plt
```
