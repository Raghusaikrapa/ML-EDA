# ILOC

```python
#it will also give the index values example
```

```python
- iloc[2:5]=====2 to 5 rows
- iloc[2:5,1:3]=====2 to 5 rows of 1 to 3 colms
- iloc[:,1:5] === all rows of 1 to 5 columns
- iloc[1:5,:]========1 to 5 rows of all columns
=====================================================================
- iloc[[0,1],[1,2,3,4,5]]  ====0,1 rows of 1,2,3,4,5, colums
- iloc[[100,200],[1,2]] ===100 & 200th rows of 1,2 colums
```

```python
df.iloc[2:5]
```

```python
df.iloc[1:3,0:3]
```

```python
df.iloc[:,1:3]
```

```python
df.iloc[0:3,:]
```

```python
df.iloc[[1,2,3],[2,3]]
```

```python
df.iloc[[100,200],[0,1]]
```
