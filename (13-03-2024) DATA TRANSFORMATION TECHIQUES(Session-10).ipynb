# Data Transformation Techniques

- we use this techniques if data do not follow Normal Distribution.
- as we know that all the maths developed is based on normal distribution only.
- it reduces skewness also
- their are some techniques
  - 1) log transformation
  - 2) expoentional transformation
  - 3) receprocial transformation
  - 4) squareroot transformation
  - 5) power transformation

```python
# import the packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the data
```

# STEP-1
# Exponentional Data

```python
# here we use numpy package
exp_bar=np.random.exponential(size=1000)
exp_bar[:10]
```

```python
# now lets plot thus values
plt.hist(exp_bar,bins=20,label='exponentional')
plt.legend()
plt.show()
```

# STEP-2
# Normal Data

```python
#lets take some normal data
nor_bar=np.random.normal(size=1000)
nor_bar[:10]
```

```python
# lets plot a histgram
plt.hist(nor_bar,bins=20,label='normal')
plt.legend()
plt.show()
```

# STEP-3
# LOG TRANSFORMATION

- np.log() is used
- log transformation do not convert data into normal
- it just reduce the skewness

```python
log=np.log(exp_bar)
log[:10]
```

```python
plt.figure(figsize=(8,4))
plt.hist(log,bins=20,label='log hist')
plt.legend()
plt.show()
```

- now lets compare the loghist and exponentinal data

```python
plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.hist(exp_bar,bins=20,label='exponentional')
plt.legend()
plt.subplot(1,2,2)
plt.hist(log,bins=20,label='log hist')
plt.legend()
plt.show()
```

# STEP-4

# RESIPROCIAL TRANSFORMATION

```python
res_hist=np.reciprocal(exp_bar)
```

```python
res_hist[:10]
```

```python
# lets plot by histgram
plt.figure(figsize=(8,4))
plt.hist(res_hist,bins=20,label='resprocal exp')
plt.legend()
plt.show()
```

```python
# lets compared the expontional and resprocaldata
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.hist(exp_bar,bins=20,label='exponentional')
plt.legend()
plt.subplot(1,2,2)
plt.hist(res_hist,bins=20,label='resprocal exp')
plt.legend()
plt.show()
```

# STEP-5

- SQUAREROOT TRANSFORMATION

```python
sqr_hist=np.sqrt(exp_bar)
```

```python
sqr_hist[:10]
```

```python
#lets draw a barplot
plt.figure(figsize=(8,3))
plt.hist(sqr_hist,bins=20,label='sqr_exp')
plt.legend()
plt.show()
```

- now lets compared both expdata and squareroot data

```python
plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.subplot(1,2,1)
plt.hist(exp_bar,bins=20,label='exponentional')
plt.legend()
plt.subplot(1,2,2)
plt.hist(sqr_hist,bins=20,label='sqr_exp')
plt.legend()
plt.show()
```

# STEP-6

- POWER TRANSFORMATION

- IT IS AVALIBLE IN SKLEAN PACKAGE
- IMPORT PACKAGE
- SAVE PACKAGE
- APPLY FITTRANSFORM

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
```

```python
power_hist=pt.fit([exp_bar]).transform([exp_bar])
power_hist[:10]
```

- lets draw histofram

```python
plt.hist(power_hist[0],bins=20)
plt.show()
```
